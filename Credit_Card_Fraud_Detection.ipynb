```python
# Feature Creation
import numpy as np
data['Hour'] = (data['Time'] // 3600) % 24
data['Log_Amount'] = np.log1p(data['Amount'])
```

```python
data.drop(['Time', 'Amount'], axis=1, inplace=True)
```

```python
# Finalized outlier columns
outlier_cols = ['V3', 'V4', 'V9', 'V10', 'V11', 'V12', 'V14', 'V16', 'V17', 'V18', 'V19']

# Add binary outlier flags
for col in outlier_cols:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    data[f'{col}_outlier'] = ((data[col] < lower_bound) | (data[col] > upper_bound)).astype(int)
```

```python
class_col = data.pop('Class')
data['Class'] = class_col
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
# Separate features and target
X = data.drop('Class', axis=1)
y = data['Class']
```

```python
# Stratified split to maintain class balance in both sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

```python
# Only scale numerical (continuous) features, not binary outlier flags
from sklearn.preprocessing import StandardScaler

scale_cols = ['Log_Amount', 'Hour'] + [col for col in data.columns if col.startswith('V') and not col.endswith('_outlier')]

# Initialize and apply scaler
scaler = StandardScaler()

X_train_scaled=X_train.copy()
X_test_scaled = X_test.copy()

# Apply scaling
X_train_scaled[scale_cols] = scaler.fit_transform(X_train[scale_cols])
X_test_scaled[scale_cols] = scaler.transform(X_test[scale_cols])
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
```

```python
# Define a reusable evaluation function
def evaluate_model(name, model, X_train, y_train, X_test, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1]

    print(f"\n========== {name} ==========")
    print(f"Accuracy:  {accuracy_score(y_test, y_pred):.4f}")
    print(f"Precision: {precision_score(y_test, y_pred, zero_division=0):.4f}")
    print(f"Recall:    {recall_score(y_test, y_pred):.4f}")
    print(f"F1 Score:  {f1_score(y_test, y_pred):.4f}")
    print(f"AUC Score: {roc_auc_score(y_test, y_proba):.4f}")
```

```python
# Initialize models
lr_model_natural = LogisticRegression(max_iter=1000, random_state=42)
dt_model_natural = DecisionTreeClassifier(random_state=42)
rf_model_natural = RandomForestClassifier(n_estimators=100, random_state=42)
xgb_model_natural = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)

# Logistic Regression â†’ on scaled features
evaluate_model("Logistic Regression", lr_model_natural, X_train_scaled, y_train, X_test_scaled, y_test)

# Others â†’ on raw features
evaluate_model("Decision Tree", dt_model_natural, X_train, y_train, X_test, y_test)
evaluate_model("Random Forest", rf_model_natural, X_train, y_train, X_test, y_test)
evaluate_model("XGBoost", xgb_model_natural, X_train, y_train, X_test, y_test)
```

```python
from sklearn.metrics import precision_recall_curve, average_precision_score
import matplotlib.pyplot as plt

def plot_pr_curve(model, X_test, y_test, label):
    y_scores = model.predict_proba(X_test)[:, 1]
    precision, recall, _ = precision_recall_curve(y_test, y_scores)
    ap_score = average_precision_score(y_test, y_scores)

    plt.plot(recall, precision, label=f"{label} (AP={ap_score:.4f})")

# Plot for top 2 models
plt.figure(figsize=(8,6))
plot_pr_curve(rf_model_over, X_test, y_test, "Random Forest (Oversampled)")
plot_pr_curve(xgb_model_over, X_test, y_test, "XGBoost (Oversampled)")

plt.title("Precision-Recall Curve for Top Models")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Feeature Importance for Random Forest
import seaborn as sns
import pandas as pd

def plot_feature_importance(model, X_train, title):
    importances = model.feature_importances_
    features = pd.Series(importances, index=X_train.columns)
    top_features = features.sort_values(ascending=False).head(15)

    plt.figure(figsize=(10,6))
    sns.barplot(x=top_features.values, y=top_features.index, palette='viridis')
    plt.title(f"Top Feature Importances - {title}")
    plt.xlabel("Importance")
    plt.ylabel("Feature")
    plt.tight_layout()
    plt.show()

plot_feature_importance(rf_model_over, X_train_over, "Random Forest (Oversampled)")
```

```python
import numpy as np
import pandas as pd

def preprocess_input(user_input):
    """
    user_input: List of 30 values [Time, V1...V28, Amount]
    Returns: Preprocessed DataFrame ready for model prediction
    """

    # Column names
    columns = ['Time'] + [f'V{i}' for i in range(1, 29)] + ['Amount']

    # Convert user input to DataFrame
    input_df = pd.DataFrame([user_input], columns=columns)

    # Step 1: Create Hour and Log_Amount features
    input_df['Hour'] = (input_df['Time'] // 3600) % 24
    input_df['Log_Amount'] = np.log1p(input_df['Amount'])

    # Step 2: Drop Time and Amount
    input_df.drop(['Time', 'Amount'], axis=1, inplace=True)

    # Step 3: Create outlier flags (as you did during training)
    outlier_cols = ['V3', 'V4', 'V9', 'V10', 'V11', 'V12', 'V14', 'V16', 'V17', 'V18', 'V19']
    for col in outlier_cols:
        input_df[f'{col}_outlier'] = ((input_df[col] > 2) | (input_df[col] < -2)).astype(int)

    # Step 4: Maintain same feature column order as during model training
    final_columns = list(input_df.columns)
    return input_df[final_columns]
```
